# 1. Generate Entailment from Entailment

Here we only consider the pair whose first sentence has only one numeric word and the second sentence has no numeric words. Generate entailment examples by the following ways.

## 1.1 Change the numeric number to other possible number
1.1.1 '2' to '3', '4'.....'10', 'two', 'three', 'four', 'five',....'ten'.
## 1.2 Change the subject in the first sentence to its hyponym

## 1.1
num1 & general $\Longrightarrow$ num1 & general 

```python
def numerical_to_general_pairs(sentence_pairs, num_word_to_replace):
    num_list = ['two','three','four','five','six','seven','eight','nine','ten','2','3','4','5','6','7','8','9','10']
    entailment_num_to_general_example = []
    for num in num_list:
        if str(num_word_to_replace.lower()) == num:
            continue
        entailment_num_to_general_example.append(
            (
                utils.return_sent_words_replaced_case_sensitive(
                    sentence_pairs[0],
                    num_word_to_replace,
                    num.capitalize() if num_word_to_replace.isupper() else num
                ),
                sentence_pairs[1]
            )
        )
    return entailment_num_to_general_example
```

```python
def generate_numerical_to_general_plural(binary_list, label, dataset_name = None):
    res = []
    for i in range(len(binary_list)):
        try:
            orig_rem_bin_pair = utils.remove_parenthesis(binary_list[i])
        except AssertionError:
            print('The ', i, ' example of ', dataset_name, 'is single-word sentence.')
            continue
        new_bin_pairs = numerical_to_general_pairs(binary_list[i], numerical_word_in_sentence(binary_list[i][0])[0])
#         new_bin_pairs = utils.same_to_same_plural_number(
#             binary_list[i],
#             numerical_word_in_sentence(binary_list[i][0])[0]
#         )
        for pair in new_bin_pairs:
            rem_bin_pair = utils.remove_parenthesis(pair)
            tmp_dict = dict()
            tmp_dict['gold_label'] = label
            tmp_dict['sentence1'] = rem_bin_pair[0]            
            tmp_dict['sentence2'] = rem_bin_pair[1]            
            tmp_dict['sentence1_binary_parse'] = pair[0]            
            tmp_dict['sentence2_binary_parse'] = pair[1]   
            tmp_dict['orig_sentence1'] = orig_rem_bin_pair[0]
            tmp_dict['orig_sentence2'] = orig_rem_bin_pair[1]
            tmp_dict['orig_sentence1_binary_parse'] = binary_list[i][0]
            tmp_dict['orig_sentence2_binary_parse'] = binary_list[i][1]
            res.append(tmp_dict)
    return res
```

```python
# Find the word to replace to an antonym word. Pattern: the word after numerical word
def return_next_word(word, sentence):
    list_of_words = sentence.split()
    if list_of_words.index(word) + 1 < len(list_of_words):
        return list_of_words[list_of_words.index(word) + 1]
    else:
        return None
        #raise Exception('{} is the last word in the sentence'.format(word))
        
noun_to_replace = return_next_word("two", "There are two boys")
print(noun_to_replace)
```

```python
from nltk.corpus import wordnet as wn
import inflect
def return_hyponym_plural(word):
    engine = inflect.engine()
    hyponym = []
    try:
        word = wn.synsets(word)[0]
    except IndexError:
        print('Hyponym of {} could not be found.'.format(word))
        return None
    hypowords = set([i for i in word.closure(lambda s:s.hyponyms())])
    for w in list(hypowords):
        h = w.lemmas()[0].name()
        h = " ".join(h.split('_'))
        hyponym.append(engine.plural(h.lower()))
        
    if len(hyponym)>0:
        return hyponym
    else:
        print('Hyponym of {} could not be found.'.format(word))
        return None
def return_hypernym_plural(word):
    engine = inflect.engine()
    hypernym = []
    try:
        word = wn.synsets(word)[0]
    except IndexError:
        print('Hypernym of {} could not be found.'.format(word))
        return None
    hyperwords = set([i for i in word.closure(lambda s:s.hypernyms())])
    for w in list(hyperwords):
        h = w.lemmas()[0].name()
        h = " ".join(h.split('_'))
        hypernym.append(engine.plural(h.lower()))
        
    if len(hypernym)>0:
        return hypernym
    else:
        print('Hypernym of {} could not be found.'.format(word))
        return None
```

```python
def replace_sub1_for_num_to_general_pairs(sentence_pairs, word_to_replace, hyponym=True):
    # If hyponym, change the word after num to its hyponyms
    num_to_general_example = []
#     num = numerical_word_in_sentence(sentence_pairs[0])[0]
#     word_to_replace = return_next_word(num, sentence_pairs[0])
#    print(word_to_replace)
    if hyponym:
        new_words = return_hyponym_plural(word_to_replace)
    else:
        new_words = return_hypernym_plural(word_to_replace)
    if new_words:
        for w in new_words:
            num_to_general_example.append(
                (
                    utils.return_sent_words_replaced_case_sensitive(
                        sentence_pairs[0],
                        word_to_replace,
                        w.capitalize() if word_to_replace.isupper() else w
                    ),
                    sentence_pairs[1]
                )
            )
        return num_to_general_example
    else:
        return []
```

```python
def generate_sub1_for_num_to_general_plural(binary_list, label, hyponym = True, dataset_name = None):
    res = []
    for i in range(len(binary_list)):
        try:
            orig_rem_bin_pair = utils.remove_parenthesis(binary_list[i])
        except AssertionError:
            print('The ', i, ' example of ', dataset_name, 'is single-word sentence.')
            continue
        num = numerical_word_in_sentence(orig_rem_bin_pair[0])[0]
        word_to_replace = return_next_word(num, orig_rem_bin_pair[0])
        if word_to_replace == None:
            continue
        new_bin_pairs = replace_sub1_for_num_to_general_pairs(binary_list[i], word_to_replace, hyponym)
#         new_bin_pairs = utils.same_to_same_plural_number(
#             binary_list[i],
#             numerical_word_in_sentence(binary_list[i][0])[0]
#         )
        for pair in new_bin_pairs:
            rem_bin_pair = utils.remove_parenthesis(pair)
            tmp_dict = dict()
            tmp_dict['gold_label'] = label
            tmp_dict['sentence1'] = rem_bin_pair[0]            
            tmp_dict['sentence2'] = rem_bin_pair[1]            
            tmp_dict['sentence1_binary_parse'] = pair[0]            
            tmp_dict['sentence2_binary_parse'] = pair[1]   
            tmp_dict['orig_sentence1'] = orig_rem_bin_pair[0]
            tmp_dict['orig_sentence2'] = orig_rem_bin_pair[1]
            tmp_dict['orig_sentence1_binary_parse'] = binary_list[i][0]
            tmp_dict['orig_sentence2_binary_parse'] = binary_list[i][1]
            res.append(tmp_dict)
    return res
```

# 3. Generate contradiction from entailment

Replace the subject of the first sentence to its antonym

```python
# Return the antonym of input word. Make them plural # TODO: pluralize doesn't make sense when the antonym is an adjective
from nltk.corpus import wordnet
#from pattern.en import pluralize
import inflect
# Return the antonym of input word. Make them plural # TODO: pluralize doesn't make sense when the antonym is an adjective
def return_antonym_plural(word):
    engine = inflect.engine()
    antonyms = []
    for syn in wn.synsets(word):
        for lm in syn.lemmas():
            if lm.antonyms():
                antonyms.append(lm.antonyms()[0].name())
                
    if antonyms:
        antonyms.sort(key=lambda item: (len(item), item))
        return engine.plural(antonyms[0])
    else:
        print('Antonym of {} could not be found.'.format(word))
        return None

return_antonym_plural('boys')
```

```python
def generate_contra_num2general_plural(binary_list):
    res = []
    for i in range(len(binary_list)):
        sentence1_bin = binary_list[i][0]
        sentence2_bin = binary_list[i][1]
        sentence1, sentence2 = utils.remove_parenthesis(binary_list[i])
        numerical_word = numerical_word_in_sentence(binary_list[i][0])[0]
        noun_to_replace = return_next_word(numerical_word, sentence1) # subject to be replaced in sentence2
        ant_noun_to_replace = return_antonym_plural(noun_to_replace) # antonym of the subject
        if ant_noun_to_replace:
            new_sentence1_bin = utils.return_sent_words_replaced_case_sensitive(
                sentence1_bin, 
                noun_to_replace, 
                ant_noun_to_replace
            )
            new_sentence1 = utils.return_sent_words_replaced_case_sensitive(
                sentence1, 
                noun_to_replace, 
                ant_noun_to_replace
            )            

            tmp_dict = dict()
            tmp_dict['gold_label'] = 'contradiction'
            tmp_dict['sentence1'] = new_sentence1
            tmp_dict['sentence2'] = sentence2
            tmp_dict['sentence1_binary_parse'] = new_sentence1_bin
            tmp_dict['sentence2_binary_parse'] =  binary_list[i][1]
            tmp_dict['orig_sentence1'] = sentence1
            tmp_dict['orig_sentence2'] = sentence2
            tmp_dict['orig_sentence1_binary_parse'] = binary_list[i][0]
            tmp_dict['orig_sentence2_binary_parse'] = binary_list[i][1]  
            res.append(tmp_dict)
        
    return res
```
