```python
def generate_embeddings(start_vec,end_vec,start_idx,end_idx):
    interpolated_values = []
    for i, each in enumerate(start_vec):
        interpolated_values.append(np.linspace(each, end_vec[i], end_idx-start_idx+1))
    np_interpolated_values = np.array(interpolated_values)
    
    result_dict = dict()
    for i in range(end_idx-start_idx+1):
        result_dict[i+start_idx] = np_interpolated_values[:,i]
    return result_dict
    



a = generate_embeddings(vec_0,vec_98,0,98)
```
